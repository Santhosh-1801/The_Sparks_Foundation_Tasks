```python
py.figure(figsize=(10,5))
sns.scatterplot(data=df1, s=100, alpha=0.7)
py.grid()
py.show()
```

```python
#Extracting the data and classifying dependent and independent values
X=df1.iloc[:,:-1]
Y=df1.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
#training testing and splitting the available data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=25) 
```

```python
#Decision Tree Algorithm
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier( criterion="entropy" , max_depth = 5 )
classifier.fit(x_train,y_train)
```

```python
y_prediction=classifier.predict(x_test)
y_prediction
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test,y_prediction)
```

```python
from sklearn.metrics import confusion_matrix
matrix=confusion_matrix(y_test,y_prediction)
matrix
```

```python
from io import StringIO
import pydotplus
import matplotlib.image as mpimg
from sklearn import tree
from sklearn.tree import export_graphviz
```

```python
text_rep=tree.export_text(classifier)
print(text_rep)
```

```python
d_data=StringIO()
filename="iris_Classify.png"
features=df1.columns.tolist()[:-1]
target=df1['Species'].unique()
fig=py.figure(figsize=(15,15))
dt=tree.export_graphviz(classifier,feature_names=features,out_file=d_data,class_names=target,filled=True,rotate=False,rounded=True,special_characters=True)
graph=pydotplus.graph_from_dot_data(d_data.getvalue())
graph.write_png(filename)
img=mpimg.imread(filename)
py.figure(figsize=(100,200),dpi=100)
py.imshow(img,interpolation='nearest')
```
